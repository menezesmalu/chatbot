```python
from elasticsearch import Elasticsearch
import warnings
import json
```

```python
warnings.filterwarnings("ignore")
```

```python
es = Elasticsearch(HOST="http://localhost", PORT=9200)
es = Elasticsearch()
```

```python
# Creating index
es.indices.create(index="movies_index")
#es.indices.delete(index="movies_index")
#es.indices.exists(index="movies_index")
```

```python
# Get data from file
f = open('result.json', 'r')
data = json.load(f)
```

```python
# Index data in elasticsearch
id_counter = 0
for q in data:
    es.index(index="movies_index", doc_type="movie_qa", id=id_counter, body=q)
    id_counter += 1 
```

```python
# Test request
# This will look for all the files in with the term "Arya" appears in the "question" field
body = {
    "from":0,
    "size":100,
    "query": {
        "match": {
            "question":"Why Arya"
        }
    }
}

res = es.search(index="movies_index", body=body)
res
```
